# Simple Linear Regression: Effect of scaler and PCA Combination

```python
X=df[['bedrooms_total',
 'baths_total',
 'acres',
 'sq_ft_tot_fn',
 'tax_gross_amount',
 'assessment_value_town',
 'garage_capacity',
 'year_built',
 'total_stories',
 'rooms_total',
 'garage',
 'flood_zone',
 'basement',
 'city__Pomfret',
 'city__Woodstock',
 'covenants__Unknown',
 'covenants__Yes',
 'garage_type',
 'short_sale',
 'basement_access_type',
 'current_use']]
y = df['price_closed']
```

```python
scaler= StandardScaler()
scaled_X = scaler.fit_transform(X)
```

After scaling, we will train a simple model using train_test_split as a starting point for comparison. Subsequently, Principal Component Analysis will be performed on the scaled set

```python
X_train,X_test,y_train,y_test= train_test_split(scaled_X,y,test_size=0.3,random_state=101)
```

```python
lm = LinearRegression()
lm.fit(X_train,y_train)
prediction = lm.predict(X_test)
print('Mean Absolute Error: {}'.format(mean_absolute_error(y_test, prediction)))
print('Mean Squared Error: {}'.format(mean_squared_error(y_test, prediction)))
print('Root Mean Squared Error: {}'.format(np.sqrt(mean_squared_error(y_test, prediction))))
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(lm, scaled_X, y, cv=10, scoring='neg_mean_squared_error')
np.mean(np.sqrt(-scores))
```

With the scores above as a starting point, I will then compare it with the score when scaling and PCA is combined. More importantly, I want to see which number of PCA features work best

```python
pca = decomposition.PCA()
df_pca = pca.fit_transform(scaled_X)
```

```python
n_params = np.arange(1,22)
```

```python
performance_PCA = []
for n in n_params:
    pipe_PCA = Pipeline([('pca', decomposition.PCA(n_components=n)),
                 ('linear', LinearRegression(normalize=True))])
    scores = cross_val_score(pipe_PCA, scaled_X, y, cv=10, scoring='neg_mean_squared_error')
    performance_PCA.append(np.mean(np.sqrt(-scores)))
```

```python
plt.plot(performance_PCA)
```

```python
print(np.min(performance_PCA))
print(np.argmin(performance_PCA)+1)
```

```python
performance_PCA = []
for n in n_params:
    pca = decomposition.PCA(n_components=n)
    X_PCA =  pca.fit_transform(X)
    X_restored = pca.inverse_transform(X_PCA)
    scores = cross_val_score(LinearRegression(), X_restored, y, cv=10, scoring='neg_mean_squared_error')
    performance_PCA.append(np.mean(np.sqrt(-scores)))
```

```python
print(np.min(performance_PCA))
print(np.argmin(performance_PCA)+1)
```

```python
plt.plot(performance_PCA)
# best performance number of features is 6
```

**Our SVR model has low bias and high variance, and so is Random Forest Regressor with high variance. We can create a stacking ensemble to see if the effect of overfitting our model will make our predictions worse off**

```python
# Manually stacking the different regression methods to get the average
estimators = []
model1 = LinearRegression(normalize=True)
estimators.append(model1)
model2 = RandomForestRegressor(max_depth=100, random_state=42,n_estimators=5)
estimators.append(model2)
model3 = SVR(kernel='linear')
estimators.append(model3)
svr_rbf = SVR(kernel='rbf')
# create the ensemble model

stregr = StackingRegressor(regressors=estimators, meta_regressor=svr_rbf)
stregr.fit(X, y)
results = cross_val_score(stregr, X, y, cv=10, scoring='neg_mean_squared_error')
np.mean(np.sqrt(-results))
```

Therefore at this stage, we can see that while more complicated, Random Forest Regressor and Support Vector Regression do not perform as well as Linear Regression. When manually stacking the models to create an ensemble, the result is worse off. We can reconfirm our doubts when running our models on the test set.

# Combine their predicting power with ensembling

I want to create an ensembling using TransformerMixin and Feature Union to aggregate the accuracy of all the model. By creating my custom class, I can also optimize model hyper parameters or pipeline. Also, it is possible to easily save and load from disk a pre-trained model.

```python
# custom transformers to pass over the transformed data
class RidgeTransformer(Ridge,TransformerMixin):
    def __init__(self, model):
        self.model = model

    def fit(self, *args, **kwargs):
        self.model.fit(*args, **kwargs)
        return self

    def transform(self, X, **transform_params):
        return pd.DataFrame(self.model.predict(X))

class LassoTransformer(Lasso, TransformerMixin):
    def __init__(self, model):
        self.model = model

    def fit(self, *args, **kwargs):
        self.model.fit(*args, **kwargs)
        return self

    def transform(self, X, **transform_params):
        return pd.DataFrame(self.model.predict(X))

class LinearRegressionTransformer(LinearRegression, TransformerMixin):
    def __init__(self, model):
        self.model = model

    def fit(self, *args, **kwargs):
        self.model.fit(*args, **kwargs)
        return self

    def transform(self, X, **transform_params):
        return pd.DataFrame(self.model.predict(X))
```

```python
def ensemble_model():
    estimators = FeatureUnion(
        transformer_list=[
            ('ridge', RidgeTransformer(Ridge(normalize=True,alpha=1e-10))),
            ('lasso', LassoTransformer(Lasso(normalize=True,alpha=0.0001))),
            ('linear',LinearRegressionTransformer(LinearRegression(normalize=True)))
        ]
    )
    # manually ensemble the above estimators to create an average
    
    ensemble = Pipeline([
        ('scaler', StandardScaler()),
        ('estimators', estimators),
        ('elastic',ElasticNet(normalize=True,alpha=1e-3,selection='random',l1_ratio=0.7)) #the one that use for prediction
    ])

    return ensemble
```

```python
# Manually create an ensemble of the different regression methods to get the average
model = ensemble_model()
scores = cross_val_score(model, X, y, cv=10, scoring='neg_mean_squared_error')
np.mean(np.sqrt(-scores))
```

```python
model.fit(X,y)
prediction = model.predict(X_Test)
print('Mean Absolute Error: {}'.format(mean_absolute_error(validate, prediction)))
print('Mean Squared Error: {}'.format(mean_squared_error(validate, prediction)))
print('Root Mean Squared Error: {}'.format(np.sqrt(mean_squared_error(validate, prediction))))
```

```python
# Using library to stack the estimators
estimators = []
model1 = LinearRegression(normalize=True)
estimators.append(model1)
model2 = Lasso(normalize=True,alpha=0.0001)
estimators.append(model2)
model3 = Ridge(normalize=True,alpha=1e-10)
estimators.append(model3)
model4 = ElasticNet(normalize=True,alpha=1e-3,selection='random',l1_ratio=0.7)
estimators.append(model4)
svr_rbf = SVR(kernel='linear')
# create the ensemble model

stregr = StackingRegressor(regressors=estimators, meta_regressor=svr_rbf)
stregr.fit(X, y)
results = cross_val_score(stregr, X, y, cv=10, scoring='neg_mean_squared_error')
np.mean(np.sqrt(-results))
```
